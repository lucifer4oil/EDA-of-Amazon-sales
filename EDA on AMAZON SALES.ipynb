```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings("ignore")

# Step 1: Load the data
df = pd.read_csv("Amazon_sales.csv")  # or pd.read_excel("Amazon_sales.xls")

# Step 2: Convert date column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Step 3: Sort and group by daily sales
df = df.sort_values('Order Date')
daily_sales = df.groupby('Order Date')['Total Sales'].sum()

# Step 4: Make it daily frequency and fill missing
daily_sales = daily_sales.asfreq('D')
daily_sales = daily_sales.fillna(0)

# Step 5: Split into train and test sets (e.g., 80% train, 20% test)
train_size = int(len(daily_sales) * 0.8)
train = daily_sales[:train_size]
test = daily_sales[train_size:]

# Step 6: Fit ARIMA model on training data
model = ARIMA(train, order=(5, 1, 0))  # You can tune (p,d,q)
model_fit = model.fit()

# Step 7: Predict the test set length
forecast = model_fit.forecast(steps=len(test))

# Step 8: Plot actual vs predicted
plt.figure(figsize=(14, 6))
plt.plot(test.index, test.values, label='Actual Sales', color='blue')
plt.plot(test.index, forecast, label='Predicted Sales', color='red')
plt.title("Actual vs Predicted Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Load the CSV file
z = "Amazon_sales.csv"  # Replace this with your actual file path
df = pd.read_csv(z)

# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by month and sum the Total Sales
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Total Sales'].sum().sort_index()
monthly_sales.index = monthly_sales.index.to_timestamp()  # For plotting

# Fit Holt-Winters Exponential Smoothing model
model = ExponentialSmoothing(
    monthly_sales,
    trend='add',
    seasonal='add',
    seasonal_periods=12,
    damped_trend=True,  # Optionally, try damped trend
    maxiter=1000  # Increase the max iterations
)
fit = model.fit()

# Forecast the next 6 months
forecast_index = pd.date_range(start=monthly_sales.index[-1] + pd.Timedelta(days=1), periods=6, freq='ME')  # Changed 'M' to 'ME'
forecast = fit.forecast(6)
forecast.index = forecast_index

# Plot the actual and forecasted data
plt.figure(figsize=(10, 8))
plt.plot(monthly_sales.index, monthly_sales.values, label='Historical Sales', marker='o')
plt.plot(forecast.index, forecast.values, label='Forecast (Next 6 Months)', marker='o', linestyle='--', color='orange')
plt.title('Monthly Total Sales with 6-Month Forecast')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Step 1: Load and prepare the data
df = pd.read_csv('Amazon_sales.csv')  # Replace with your actual file path

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by month and sum the 'Total Sales'
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Total Sales'].sum().sort_index()

# Convert period index to timestamp for easier plotting
monthly_sales.index = monthly_sales.index.to_timestamp()

# Step 2: Visualize the data (initial view of trend)
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values)
plt.title("Monthly Total Sales (Initial View)")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Step 3: Fit the ARIMA model
# (p, d, q) - p: number of lag observations, d: number of differencing, q: size of moving average window
# You can experiment with different (p, d, q) values for better forecasting

model = ARIMA(monthly_sales, order=(5, 1, 0))  # Example parameters (p=5, d=1, q=0)
fit = model.fit()

# Step 4: Forecast the next 36 months (3 years)
forecast = fit.forecast(steps=36)

# Step 5: Create a future time index for the next 36 months
forecast_index = pd.date_range(monthly_sales.index[-1], periods=37, freq='M')[1:]

# Step 6: Plot the historical data and forecast
plt.figure(figsize=(12, 8))
plt.plot(monthly_sales.index, monthly_sales.values, label='Historical Sales', marker='o')
plt.plot(forecast_index, forecast, label='Forecast (Next 3 Years)', marker='o', linestyle='--', color='orange')
plt.title('ARIMA Forecast for Monthly Total Sales with 3-Year Forecast')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Step 1: Load and prepare the data
df = pd.read_csv('Amazon_sales.csv')  # Replace with your actual file path

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by month and sum the 'Total Sales'
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Total Sales'].sum().sort_index()

# Convert period index to timestamp for easier plotting
monthly_sales.index = monthly_sales.index.to_timestamp()

# Step 2: Visualize the data (initial view of trend)
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values)
plt.title("Monthly Total Sales (Initial View)")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Step 3: Fit the model for trend analysis (no seasonal component)
model = ExponentialSmoothing(monthly_sales, trend='add', seasonal=None, seasonal_periods=12)
fit = model.fit()

# Step 4: Forecast the next 36 months (3 years) based on the trend
forecast = fit.forecast(36)

# Create a future time index for the next 36 months (Month End frequency)
forecast_index = pd.date_range(monthly_sales.index[-1], periods=37, freq='ME')[1:]

# Step 5: Plot the historical data and the trend-based forecast
plt.figure(figsize=(12, 8))
plt.plot(monthly_sales.index, monthly_sales.values, label='Historical Sales', marker='o')
plt.plot(forecast_index, forecast.values, label='Forecast (Next 3 Years)', marker='o', linestyle='--', color='orange')
plt.title('Trend Analysis for Monthly Total Sales with 3-Year Forecast')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
